#Data preprocessing

Dropping NaN values:

```python
df_dropped = df.dropna(axis=0).reset_index(drop=True)
print(df_dropped.shape)
```

```python
grouped_data = df_dropped.groupby("Course")
missing_values_by_course = grouped_data.apply(lambda x: x.isnull().sum().sum())
print("Number of missing values by course:\n", missing_values_by_course)
```

Filling Missing Values, (we have none)


```python
data_filled = df.copy()
Courses = data_filled["Course"].unique()
for course in Courses:
    mask = data_filled["Course"] == course
    data_filled.loc[mask] = data_filled.loc[mask].ffill()
    data_filled.loc[mask] = data_filled.loc[mask].bfill()
    data_filled.loc[mask] = data_filled.loc[mask].fillna(0)

print(data_filled.shape)
```

Numerical Encoding, transferring categorical variables into a numerical format

```python
print(f"Unique Status Values:\n{df_dropped['Target'].unique()}")
```

Label Encoding

```python
label_encoder = LabelEncoder()
status_label_encoded = label_encoder.fit_transform(df_dropped["Target"])

print("Label Encoder Mappings:")
for encoded_label, original_label in enumerate(label_encoder.classes_):
    print(f"{encoded_label} -> {original_label}")

print("\nLabel Encoded Data:")
print(status_label_encoded)
```

Ordinal Encoding

```python
status_map = {
    "Enrolled": 0,
    "Graduate": 1,
    "Dropout": 2,
}

df_dropped["Target_mapped"] = df_dropped["Target"].map(status_map)
print(df_dropped[["Course", "Target", "Target_mapped"]].head(20))
```

```python
def plot_cm(y_true, y_pred, cmap="Blues", figsize=(4, 4)):

   # Plot confusion matrix heatmap.

    #Parameters:
   # y_true (array-like)# True labels.
   # y_pred (array-like) # Predicted labels.
   # cmap (str, optional) # Colormap for the heatmap. Defaults to 'Blues'.

    cm = confusion_matrix(y_true, y_pred)

    plot_confusion_matrix(
        conf_mat=cm,
        show_absolute=True,
        show_normed=True,
        colorbar=True,
        cmap=cmap,
        figsize=figsize,
    )
    plt.show()
```

```python
def plot_comparison(dataframe, title="Comparison", figsize=(15, 5)):
    metrics = dataframe.columns[1:]
    algorithm_values = dataframe.iloc[0, 1:]

    # Set the positions for the bars
    x = range(len(algorithm_values))

    # Plot the bar plot
    plt.figure(figsize=figsize)
    bars = plt.bar(x, algorithm_values, color=['tab:cyan', 'tab:cyan', 'tab:cyan'])  # Colors for KNN, Tree, SVM

    # Add metric labels on top of each bar
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.2f}', ha='center', va='bottom')

    # Add labels, title, and legend
    plt.xlabel('Algorithm')
    plt.ylabel('Accuracy')
    plt.title(title)
    plt.xticks(x, metrics)

    # Show plot
    plt.tight_layout()
    plt.show()
```

```python
from hyperopt import hp, fmin, tpe, STATUS_OK, Trials
from sklearn.model_selection import cross_val_score

crt = ['absolute_error', 'squared_error']
max_d = np.append(range(3,15),None)
max_f = np.append(range(2,10), [None])
min_s_l = range(2,10)
min_s_s = range(2,10)
est = [10, 50, 100]

space = {'criterion': hp.choice('criterion', crt),
        'max_depth': hp.choice('max_depth', max_d),
        'max_features': hp.choice('max_features', max_f),
        'min_samples_leaf': hp.choice('min_samples_leaf', min_s_l),
        'min_samples_split' : hp.choice('min_samples_split', min_s_s),
        'n_estimators' : hp.choice('n_estimators', est),
        'ccp_alpha' : hp.uniform('ccp_alpha', 0, 1)
    }
def objective(space):
    model = RandomForestRegressor(criterion = space['criterion'],
                                 max_depth = space['max_depth'],
                                 max_features = space['max_features'],
                                 min_samples_leaf = space['min_samples_leaf'],
                                 min_samples_split = space['min_samples_split'],
                                 n_estimators = space['n_estimators'],
                                 ccp_alpha = space['ccp_alpha']
                                 )

    mae = cross_val_score(model, Xs_train, ys_train.values.ravel(), cv = 3,
                          scoring='neg_mean_absolute_error').mean()
    #print (mae)
    # We aim to maximize accuracy, therefore we return it as a negative value
    return {'loss': -mae, 'status': STATUS_OK }

trials = Trials()
best = fmin(fn= objective,
            space= space,
            algo= tpe.suggest,
            max_evals = 30,
            trials= trials)
best
```
